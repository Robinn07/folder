```python
# 1. Install dependencies (includes ffmpeg)
!pip install git+https://github.com/linto-ai/whisper-timestamped ffmpeg-python pandas imageio[ffmpeg]

# 2. Add ffmpeg to PATH from imageio (no system install needed)
import imageio_ffmpeg as ffmpeg
import os

os.environ["PATH"] += os.pathsep + os.path.dirname(ffmpeg.get_ffmpeg_exe())
print("✅ ffmpeg path set:", ffmpeg.get_ffmpeg_exe())

# 3. Imports
import whisper_timestamped as whisper
import pandas as pd
from pathlib import Path

# 4. Select your file (change only here)
audio_path = r"C:/Users/HP/Downloads/vad.mp3"  # works for .mp3, .wav, .mp4, etc.

# 5. Check file
if not os.path.exists(audio_path):
    raise FileNotFoundError(f"❌ File not found: {audio_path}")
print(f"🎵 Processing file: {Path(audio_path).name}")

# 6. Load model
model = whisper.load_model("small")  # 'tiny' = fastest, 'large' = most accurate

# 7. Transcribe (works for audio + video)
result = whisper.transcribe(audio_path)

# 8. Extract word timestamps
words_data = []
for segment in result['segments']:
    for word in segment['words']:
        words_data.append({
            "word": word['text'],
            "start_time": round(word['start'], 2),
            "end_time": round(word['end'], 2)
        })

# 9. Save to CSV
df = pd.DataFrame(words_data)
output_csv = Path(audio_path).stem + "_timestamps.csv"
df.to_csv(output_csv, index=False)

print(f"✅ Word timestamps saved to: {output_csv}")
df.head(10)
```

```python
import whisper
audio = r"C:\Users\HP\Downloads\vad 1-2.m4a"

audio = whisper.load_audio(audio_path)
audio = whisper.pad_or_trim(audio)

# Make log-Mel spectrogram and move to the same device as the model
mel = whisper.log_mel_spectrogram(audio).to(model.device)

# Detect the spoken language (optional)
_, probs = model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

# Decode the audio
options = whisper.DecodingOptions()
result = whisper.decode(model, mel, options)

print(result.text)
```
